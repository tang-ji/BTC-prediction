## Pull sentiment, fear greed and BTC/USDT data from bitcurate API.
Can read more about this API at https://doc.api.bitcurate.com/

```python
r = requests.get('https://datascience.api.dev.bitcurate.com/social_sentiment?query=(BTC%20OR%20bitcoin)%20AND%20binance&before_date=8/13/2020%200:0')
sentiment = r.json()
sentiment.keys()
```
